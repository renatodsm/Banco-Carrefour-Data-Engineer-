```python
#Qual foi o valor da receita total?
df['Valor Venda'].sum()


```

```python
#Qual foi o valor do custo total?
df['custo']=df['Custo Unitário'].mul(df['Quantidade'])
```

```python
df.head(50)
```

```python
#Qual o custo total?
round(df['custo']).sum()
```

```python
#Qual o custo total?
round(df['Lucro']).sum()
```

```python
#vamos criar a coluna do lucro receita -custo
df['Lucro']=df['Valor Venda'] -  df['custo']
df.head()
```

```python
#Coluna tempo envio
df['tempo envio']=(df['Data Envio'] - df['Data Venda']).dt.days
df.head()
```

```python
df.groupby('Marca')['tempo envio'].mean()
```

```python
#Verificar Valores ausentes
df.isnull().sum()
```

```python
#Agrupando por ano e por marca
df.groupby([df['Data Venda'].dt.year,'Marca']
           )['Lucro'].sum()
```

```python
pd.options.display.float_format='{:20,.2f}'.format
```

```python
#resetando o index
df.groupby([df['Data Venda'].dt.year,'Marca']
           )['Lucro'].sum().reset_index()
```

```python
#Qual o total de produtos vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
#Gráfico Total de  Produtos Vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values().plot.barh(title="Total Produto vendidos")
plt.xlabel('Total')
plt.ylabel('Produto')
```

```python
#Gráfico Lucro por Ano
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().sort_values().plot.bar(title="Lucro x Ano")
plt.xlabel('ano')
plt.ylabel('Receita')
```

```python
#Apenas as vendas de 2009
df_2009=df[df['Data Venda'].dt.year==2009]
df_2009.head()
```

```python
#Gráfico de linha Lucro por Mês
df_2009.groupby(df_2009['Data Venda'].dt.month)['Lucro'].sum().plot(title="Lucro x mês")
plt.xlabel('Mês')
plt.ylabel('Lucro');
```

```python
#Gráfico de Barras Lucro por Marca
df_2009.groupby(df_2009['Marca'])['Lucro'].sum().plot.bar(title="Lucro x Marca")
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```

```python
##Gráfico de Barras Lucro por Classe
df_2009.groupby(df_2009['Classe'])['Lucro'].sum().plot.bar(title="Lucro x Classe")
plt.xlabel('Classe')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```
